```python

def preprocessor(file):
    col_name = ['lon','lat']
    mon = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
    for i in range(1901,2022):
        for ii in range(12):
            col_name.append(str(i)+"_"+mon[ii])
    read_path = os.path.join(path,file)
    try:
        df = pd.read_csv(read_path)
    except:
        df = pd.read_excel(read_path)

    try:
        df.columns = col_name
    except:
        print("Erro While Renaming Columns")
        print("Len of Col_name", len(col_name),"len of df columns", len(list(df.columns)))
        df.rename(columns = {'Long':'lon','Lat':'lat'}, inplace = True)
        df.columns = col_name[:-2]
        print("Now the len of df columns",len(list(df.columns)))

    return df
        
```
